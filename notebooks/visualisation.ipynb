```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
import numpy as np

topics = ["machine learning, ai, deep learning", "python, flask, web", "machine learning, neural networks"]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(topics)

similarity_matrix = cosine_similarity(tfidf_matrix)
print(similarity_matrix)
```

```python
sparse_data = {
    "row_indices": tfidf_matrix.nonzero()[0].tolist(),  # Row indices of non-zero elements
    "col_indices": tfidf_matrix.nonzero()[1].tolist(),  # Column indices of non-zero elements
    "data": tfidf_matrix.data.tolist(),  # Non-zero values
    "shape": tfidf_matrix.shape  # Shape of the matrix
}

sparse_data
```

```python
# Reconstruct the csr_matrix from the stored data
row_indices = np.array(sparse_data["row_indices"])
col_indices = np.array(sparse_data["col_indices"])
data = np.array(sparse_data["data"])
shape = sparse_data["shape"]
reconstructed_csr = csr_matrix((data, (row_indices, col_indices)), shape=shape)
```
