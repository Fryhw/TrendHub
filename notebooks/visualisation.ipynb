```python
import requests
import json
import re
import base64

from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi


token_loc = "../tokens/"
data_destination = "../data/external/"
repos = "/search/repositories"

# Get username and password from .txt file (format is "<username> <password>")
with open(token_loc + "/db_credentials.txt") as file:
    db_credentials = file.read().strip()
    db_credentials = db_credentials.split(sep=" ")
    username, db_password = db_credentials
    
    

uri = "mongodb+srv://" + username +":" + db_password + "@trendhubcluster.pmiir.mongodb.net/?retryWrites=true&w=majority&appName=TrendHubCluster"
# Create a new client and connect to the server
client = MongoClient(uri, server_api=ServerApi('1'))



with open(token_loc +"/token.txt") as file:
    your_github_token = file.read().strip()

HEADERS = {"Authorization": f"Bearer {your_github_token}"}
BASE_URL = "https://api.github.com"

def get_repo_info(owner, repo):
    url = f"{BASE_URL}/repos/{owner}/{repo}"
    response = requests.get(url, headers=HEADERS)
    response.raise_for_status() 
    data = response.json()
    return data

def get_repo_issues(owner, repo):
    url = f"{BASE_URL}/repos/{owner}/{repo}/issues"
    params = {"state": "all"} 
    response = requests.get(url, headers=HEADERS, params=params)
    response.raise_for_status()
    data = response.json()
    return data

def get_repo_contributors(owner, repo):
    url = f"{BASE_URL}/repos/{owner}/{repo}/contributors"
    response = requests.get(url, headers=HEADERS)
    response.raise_for_status()
    data = response.json()
    return data


def get_repo_readme(owner, repo):
    url = f"{BASE_URL}/repos/{owner}/{repo}/readme"
    response = requests.get(url, headers=HEADERS)
    response.raise_for_status()
    data = response.json()
    return data

def base64_to_string(base):
    return base64.b64decode(base).decode('utf-8')


owner = "octocat"
repo = "Hello-World"


owner ="Fryhw"
repo = "TrendHub"

repo_info = get_repo_info(owner, repo)

issues = get_repo_issues(owner, repo)

contributors = get_repo_contributors(owner, repo)

readme = get_repo_readme(owner, repo)

with open(data_destination +"repo_info.json", "w") as f:
    json.dump(repo_info, f, indent=4)

with open(data_destination + "issues.json", "w") as f:
    json.dump(issues, f, indent=4)

with open(data_destination + "contributors.json", "w") as f:
    json.dump(contributors, f, indent=4)

with open(data_destination+ "readme.json", "w") as f:
    json.dump(readme, f, indent=4)

readme_content = base64_to_string(readme['content'])
title = repo_info['name']
description = repo_info['description']
print(readme_content)


# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)

client.get_database('sample_mflix').get_collection('users').find_one()
```
